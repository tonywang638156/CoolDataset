```python
import pandas as pd
df = pd.read_csv("/content/dog-breeds-ranking-best-to-worst/dogs-ranking-dataset.csv")
df.head(3)
```

## Size

图片.png

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_clean = df[['popularity ranking', 'size', 'intelligence', 'LONGEVITY(YEARS)', 'PURCHASE PRICE']].dropna()

# Bar chart: Size vs Popularity Ranking
plt.figure(figsize=(6, 3))
sns.barplot(x='size', y='popularity ranking', data=df_clean, palette='Blues')
plt.title('Size vs Popularity Ranking')
plt.xlabel('Size (1=small, 2=medium, 3=large)')
plt.ylabel('Popularity Ranking')
plt.grid(True)
plt.show()
```

图片.png

Small Size Dogs Win!

```python
# Define the desired order for the intelligence levels
intelligence_order = ['Lowest', 'Fair', 'Average', 'Above average', 'Excellent', 'Brightest']

# Create the bar chart with the specified order
plt.figure(figsize=(7, 3))
sns.barplot(x='intelligence', y='popularity ranking', data=df_clean, palette='coolwarm', order=intelligence_order)
plt.title('Intelligence vs Popularity Ranking')
plt.xlabel('Intelligence (Lowest to Brightest)')
plt.ylabel('Popularity Ranking')
plt.grid(True)
plt.show()
```

图片.png


Smart Dogs Win!

## Longevity & Price
图片.png

图片.png


```python
import matplotlib.pyplot as plt

# Modify the code to display the subplots side by side (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(10, 3))  # 1 row, 2 columns

# Scatter plot: Longevity (Years) vs Popularity Ranking
axes[0].scatter(df_clean['LONGEVITY(YEARS)'], df_clean['popularity ranking'], color='orange')
axes[0].set_title('Longevity (Years) vs Popularity Ranking')
axes[0].set_xlabel('Longevity (Years)')
axes[0].set_ylabel('Popularity Ranking')
axes[0].grid(True)

# Scatter plot: Purchase Price vs Popularity Ranking
axes[1].scatter(df_clean['PURCHASE PRICE'], df_clean['popularity ranking'], color='purple')
axes[1].set_title('Purchase Price vs Popularity Ranking')
axes[1].set_xlabel('Purchase Price ($)')
axes[1].set_ylabel('Popularity Ranking')
axes[1].grid(True)

# Adjust layout for clarity and spacing
plt.tight_layout()

# Display the plots
plt.show()
```
